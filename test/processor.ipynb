```python
import pandas as pd

df = pd.read_html('courses.html')
```

```python
df = df[0]
```

```python
df['Level'] = df['Course'].apply(lambda x: int(x.split(' ')[0].split('-')[1][::-1]) % 10)
```

```python
df
```

```python
out = ''

for l in [4, 3, 2, 1]:
    out += f'## Level-{l}\n\n'
    temp = df[df['Level'] == l]
      
    for i in temp.itertuples():
        # print(i)
        out += f'''
                ### {i.Course}
                
                Institute: {i.Institute}
                
                {i._3}
              '''
        
```

```python
print(out)
```
